```python
import pandas as pd 
import numpy as np
import os

import cv2
import matplotlib.pylab as plt
```

```python

def weight(folder_in):
    tamano=[]
    archivos = os.listdir(folder_in)
    for archivo in archivos:
        ruta_entrada = os.path.join(folder_in, archivo)
        tamano_kb = os.path.getsize(ruta_entrada) / 1024
        #if tamano_kb >=18:
            #crear carpeta que contenga solo las images con información relevante
            
        tamano.append(tamano_kb)
    return tamano
    

folder_in = r"/home/catalina/Escritorio/PAPERS MEMORIA/Glaciar"
tamano=weight(folder_in)
```

```python
print(len(tamano))
pd.Series(tamano).plot(kind='hist', bins=25)

plt.show()
```

```python
k=0
for i in tamano:
    if i>=18:
        k+=1
    i+=1  

print(k)
```
